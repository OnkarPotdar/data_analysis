```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,StratifiedKFold,GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.ensemble import VotingClassifier
```

```python
cancer = pd.read_csv("D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Wisconsin/BreastCancer.csv",index_col=0)
```

```python
cancer
```

```python
dum_cancer = pd.get_dummies(cancer, drop_first=True)
```

```python
dum_cancer
```

```python
X = dum_cancer.drop('Class_Malignant', axis=1)
y = dum_cancer['Class_Malignant']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                              stratify=y, random_state=2023)
```

```python
clf = DecisionTreeClassifier(random_state=2023)
nb = GaussianNB()
svm = SVC(probability=True, random_state=2023, kernel='linear')
```

```python
models = [('Tree', clf),('Naive', nb),('SVM', svm)]
```

```python
voting = VotingClassifier(models, voting='soft')
voting.fit(X_train, y_train)
```

```python
y_pred = voting.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
y_pred_prob = voting.predict_proba(X_test)[:,1]
print(roc_auc_score(y_test, y_pred_prob))
```
