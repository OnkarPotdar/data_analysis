```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
```

```python
b_cancer=pd.read_csv("D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Wisconsin/BreastCancer.csv")
```

```python
b_cancer
```

```python
dum_b_cancer = pd.get_dummies(b_cancer, drop_first=True)
X = dum_b_cancer.drop('Class_Malignant',axis=1)
y = dum_b_cancer['Class_Malignant']
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=2023)
knn = KNeighborsClassifier(n_neighbors=3)
results = cross_val_score(knn, X, y ,scoring='roc_auc',
                          cv = kfold)
print(results)
print(results.mean())
```

```python
roc = []
ks = np.arange(1,30,2)
for i in ks:
    knn = KNeighborsClassifier(n_neighbors=i)
    results = cross_val_score(knn, X, y ,scoring='roc_auc',
                              cv = kfold)
    roc.append(results.mean())

i_max = np.argmax(roc)
print("Best k =", ks[i_max], "with score =", np.max(roc)) 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
b_cancer=pd.read_csv('D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Wisconsin/BreastCancer.csv')
```

```python
b_cancer
```

```python
dum_b_cancer = pd.get_dummies(b_cancer, drop_first=True)
X = dum_b_cancer.drop('Class_Malignant',axis=1)
y = dum_b_cancer['Class_Malignant']
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=2023)
knn = KNeighborsClassifier()
```

```python
params = {'n_neighbors':np.arange(1,30,2)}
gcv = GridSearchCV(knn, param_grid=params,scoring='roc_auc',
                          cv = kfold, verbose=3)
```

```python
gcv.fit(X, y)
print(gcv.best_params_)
print(gcv.best_score_)

pd_gcv = pd.DataFrame(gcv.cv_results_)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
imag_seg = pd.read_csv('D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Image Segmentation/Image_Segmention.csv')
```

```python
imag_seg
```

```python
X = imag_seg.drop('Class', axis=1)
y = imag_seg['Class']
```

```python
le = LabelEncoder()
le_y = le.fit_transform(y)
print(le.classes_)
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=2023)
```

```python
scaler = StandardScaler()
knn = KNeighborsClassifier()
pipe = Pipeline([('scl_std', scaler), ('knn_model', knn)])
```

```python
params = {'knn_model__n_neighbors':np.arange(1,31)}
gcv = GridSearchCV(pipe, param_grid=params,scoring='neg_log_loss',
                          cv = kfold)
```

```python
gcv.fit(X, y)
print(gcv.best_params_)
print(gcv.best_score_)

pd_gcv = pd.DataFrame(gcv.cv_results_)
```
