```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import roc_auc_score
```

```python
telecom=pd.read_csv("D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Telecom/Telecom.csv")
```

```python
telecom
```

```python
dum_tel = pd.get_dummies(telecom,drop_first=True)
```

```python
dum_tel
```

```python
X=dum_tel.drop('Response_Y',axis=1)
y=dum_tel['Response_Y']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2023,
                                              test_size=0.30,stratify=y)
```

```python
nb=BernoulliNB()
nb.fit(X_train,y_train)
```

```python
pred=nb.predict(X_test)
```

```python
pred
```

```python
y_pred_prob = nb.predict_proba(X_test)[:,1]
```

```python
y_pred_prob
```

```python
print(roc_auc_score(y_test, y_pred_prob))
```
