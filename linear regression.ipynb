```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
load_boston()  # this will shows the  information in key value format.
```

```python
# to create dataframe
df=load_boston()
```

```python
df
```

```python
# to createe in tabluar formaat
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df['feature_names']
```

```python
dataset
```

```python
# independant features and dependent features
X=dataset
y=df.target
```

```python
X
```

```python
y
```

```python
# we can apply train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                test_size=0.33, random_state=42)
```

```python
X_train
```

```python
# standardizing the model for better model
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train) # it will transform the X-train data
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
regression.fit(X,y)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',
                    cv=5)
```

```python
np.mean(mse)
```

```python
# the mse is always small beacause the mse is thee diff between predicted value
# and actual value. hence it will be less
```

```python
# prediction
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
```

```python
sns.displot(reg_pred-y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_pred,y_test)
```
