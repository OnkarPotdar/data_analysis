```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, KFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
```

```python
concrete=pd.read_csv("D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Concrete Strength/Concrete_Data.csv")
```

```python
concrete
```

```python
X=concrete.drop('Strength',axis=1)
y=concrete['Strength']
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=2023)
```

```python
scaler=StandardScaler()
knn=KNeighborsRegressor()
pipe=Pipeline([('sclaer_model',scaler),('Knn_model',knn)])
print(pipe.get_params())
```

```python
params={'Knn_model__n_neighbors':np.arange(1,31)}
gcv=GridSearchCV(pipe,param_grid=params,scoring='r2',cv=kfold)
```

```python
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
pd_gcv = pd.DataFrame(gcv.cv_results_)
pd_gcv
```

```python
from sklearn.impute import SimpleImputer
```

```python
chem = pd.read_csv("D:/onkar_cdac/C-DAC/PML/PML/Files_by_sir/5. PML/1. Day wise Study Material/Cases/Chemical Process Data/ChemicalProcess.csv")
```

```python
chem
```

```python
X=chem.drop('Yield',axis=1)
y=chem['Yield']
```

```python
imputer = SimpleImputer()
scaler = StandardScaler()
knn = KNeighborsRegressor()
```

```python
pipe = Pipeline([('imputing',imputer),('scaling', scaler),
                 ('knn_model',knn)])
```

```python
print(pipe.get_params())
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=2023)
```

```python
params = {'knn_model__n_neighbors':np.arange(1,31),
          'imputing__strategy':['mean','median']}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,scoring='r2',cv=kfold,verbose=3)
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)

pd_gcv = pd.DataFrame(gcv.cv_results_)
```

```python
pd_gcv
```
